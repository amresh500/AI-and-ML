```python
# Task 1

def perform_unit_conversion(mode, input_value):
    """
    Transforms a numeric value from one unit of measurement to another.

    Parameters:
    mode (int): The conversion mode chosen by the user.
        1 - Meters to Feet
        2 - Feet to Meters
        3 - Kilograms to Pounds
        4 - Pounds to Kilograms
        5 - Liters to Gallons
        6 - Gallons to Liters
    input_value (float): The numeric quantity to be converted.

    Returns:
    float: The resulting converted quantity.
    """

    if mode == 1:
        return input_value * 3.28084  # meters to feet
    elif mode == 2:
        return input_value / 3.28084  # feet to meters
    elif mode == 3:
        return input_value * 2.20462  # kg to pounds
    elif mode == 4:
        return input_value / 2.20462  # pounds to kg
    elif mode == 5:
        return input_value * 0.264172  # liters to gallons
    elif mode == 6:
        return input_value / 0.264172  # gallons to liters
    else:
        return None


def run_converter():
    print("==== Unit Converter ====")
    print("Select Conversion Type:")
    print("1. Meters to Feet")
    print("2. Feet to Meters")
    print("3. Kilograms to Pounds")
    print("4. Pounds to Kilograms")
    print("5. Liters to Gallons")
    print("6. Gallons to Liters")

    try:
        selected_option = int(input("Enter your choice (1-6): "))

        if selected_option < 1 or selected_option > 6:
            print("Error: Invalid conversion type selected.")
            return

        user_value = float(input("Enter the value to convert: "))

        converted = perform_unit_conversion(selected_option, user_value)

        if converted is not None:
            print(f"Converted value: {converted:.4f}")
        else:
            print("Error: Conversion could not be performed.")

    except ValueError:
        print("Error: Please enter a valid numeric value.")


if __name__ == "__main__":
    run_converter()
```

```python
def compute_total(data):
    """
    Computes the total sum of all values in the data list.

    Parameters:
    data (list of float): The collection of numeric values.

    Returns:
    float: The total sum of the values.
    """
    return sum(data)


def compute_mean(data):
    """
    Computes the arithmetic mean of all values in the data list.

    Parameters:
    data (list of float): The collection of numeric values.

    Returns:
    float: The mean of the values.
    """
    return sum(data) / len(data)


def get_largest(data):
    """
    Retrieves the largest value from the data list.

    Parameters:
    data (list of float): The collection of numeric values.

    Returns:
    float: The highest number in the list.
    """
    return max(data)


def get_smallest(data):
    """
    Retrieves the smallest value from the data list.

    Parameters:
    data (list of float): The collection of numeric values.

    Returns:
    float: The lowest number in the list.
    """
    return min(data)


def run_stats_menu():
    print("==== Mathematical Operations on a List ====")
    print("Select Operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        selected_option = int(input("Enter your choice (1-4): "))

        if selected_option < 1 or selected_option > 4:
            print("Error: Invalid operation selected.")
            return

        raw_input = input("Enter numbers separated by spaces: ")

        # Convert input string to list of floats
        data = list(map(float, raw_input.split()))

        if len(data) == 0:
            print("Error: The list cannot be empty.")
            return

        if selected_option == 1:
            output = compute_total(data)
            print(f"Sum: {output}")
        elif selected_option == 2:
            output = compute_mean(data)
            print(f"Average: {output}")
        elif selected_option == 3:
            output = get_largest(data)
            print(f"Maximum: {output}")
        elif selected_option == 4:
            output = get_smallest(data)
            print(f"Minimum: {output}")

    except ValueError:
        print("Error: Please enter valid numeric values.")


if __name__ == "__main__":
    run_stats_menu()
```

**Exercise on Nested List:**

```python
# 1. Collapse a Nested List into a Single Layer:

def collapse_nested(nested):
    collapsed = []
    for inner in nested:
        for val in inner:
            collapsed.append(val)
    return collapsed


# Example usage
grouped_nums = [[1, 2], [3, 4], [5]]
output = collapse_nested(grouped_nums)
print(output)
```

```python
# 2. Retrieve Element by Index Path in a Nested List:

def fetch_by_path(nested, path):
    target = nested
    for idx in path:
        target = target[idx]
    return target

# Example usage:
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
index_path = [1, 2]

output = fetch_by_path(matrix, index_path)
print(output)
```

```python
# 3. Compute the Total of All Values in a Nested List:

def total_nested(nested):
    running_total = 0
    for entry in nested:
        if isinstance(entry, list):
            running_total += total_nested(entry)  # Recursively accumulate nested list values
        else:
            running_total += entry
    return running_total

# Example usage:
mixed_nested = [[1, 2], [3, [4, 5]], 6]
print(total_nested(mixed_nested))  # Output: 21
```

```python
# 4. Delete a Target Value from a Nested List:

def delete_target(nested, target):
    filtered = []
    for entry in nested:
        if isinstance(entry, list):
            filtered.append(delete_target(entry, target))
        elif entry != target:
            filtered.append(entry)
    return filtered

# Example usage:
grouped = [[1, 2], [3, 2], [4, 5]]
target_val = 2
print(delete_target(grouped, target_val))
```

```python
# 5. Locate the Peak Element in a Nested List:

def locate_peak(nested):
    peak = float('-inf')  # Begin with the lowest possible starting value
    for entry in nested:
        if isinstance(entry, list):
            peak = max(peak, locate_peak(entry))  # Recursively locate peak in sublist
        else:
            peak = max(peak, entry)  # Compare against current entry
    return peak

# Example usage:
mixed_nested = [[1, 2], [3, [4, 5]], 6]
print(locate_peak(mixed_nested))
```

```python
# 6. Tally Occurrences of a Value in a Nested List:
def tally_occurrences(nested, target):
    tally = 0
    for entry in nested:
        if isinstance(entry, list):
            tally += tally_occurrences(entry, target)  # Recursively tally in sublist
        elif entry == target:
            tally += 1
    return tally

# Example usage:
grouped = [[1, 2], [2, 3], [2, 4]]
target_val = 2
print(tally_occurrences(grouped, target_val))
```

```python
# 7. Recursively Flatten a Deeply Nested List:

def fully_flatten(nested):
    flat = []
    for entry in nested:
        if isinstance(entry, list):
            flat.extend(fully_flatten(entry))  # Recursively flatten all sublists
        else:
            flat.append(entry)
    return flat

# Example usage:
deep_nested = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(fully_flatten(deep_nested))
```

```python
# 8. Mean Value of All Elements in a Nested List:

def mean_of_nested(nested):
    def accumulate(nested):
        grand_total = 0
        item_count = 0
        for entry in nested:
            if isinstance(entry, list):
                sub_total, sub_count = accumulate(entry)
                grand_total += sub_total
                item_count += sub_count
            else:
                grand_total += entry
                item_count += 1
        return grand_total, item_count

    grand_total, item_count = accumulate(nested)
    return grand_total / item_count if item_count != 0 else 0

# Example usage:
grouped_nums = [[1, 2], [3, 4], [5, 6]]
print(mean_of_nested(grouped_nums))
```

```python
# Problem - 4: Matrix Operations:

# Given Matrices
M = np.array([[3, 4],
              [7, 8]])

N = np.array([[5, 3],
              [2, 1]])

# 1. Prove M.M−1 = I

M_inv = np.linalg.inv(M)       # Inverse of M
identity_check = np.matmul(M, M_inv)   # M * M⁻¹

print("M inverse:\n", M_inv)
print("\nM * M⁻¹:\n", identity_check)
```

```python
# Demonstrate MN ≠ NM

product_MN = np.matmul(M, N)
product_NM = np.matmul(N, M)

print("MN:\n", product_MN)
print("\nNM:\n", product_NM)
```

```python
# Verify (MN)^T = N^T M^T

lhs = (np.matmul(M, N)).T
rhs = np.matmul(N.T, M.T)

print("(MN)^T:\n", lhs)
print("\nN^T M^T:\n", rhs)
```

```python
# Solve a System of Linear Equations Using Matrix Form

# Represent in Matrix Form CoeffMatrix * VarVector = ConstVector
CoeffMatrix = np.array([[2, -3, 1],
                        [1, -1, 2],
                        [3,  1, -1]])

ConstVector = np.array([-1, -3, 9])

CoeffMatrix_inv = np.linalg.inv(CoeffMatrix)
SolutionVec = np.matmul(CoeffMatrix_inv, ConstVector)

print("Solution using inverse method:\n", SolutionVec)
print(f"x = {SolutionVec[0]}\ny = {SolutionVec[1]}\nz = {SolutionVec[2]}")
```

```python
# Solve Directly Using np.linalg.solve()

DirectSolution = np.linalg.solve(CoeffMatrix, ConstVector)
print("Solution using np.linalg.solve():\n", DirectSolution)
print(f"x = {DirectSolution[0]}\ny = {DirectSolution[1]:.1f}\nz = {DirectSolution[2]}")
```

```python
import time
import random

# Element-wise Addition using Python lists:

# Generate two lists each containing 1,000,000 random values
data_list_a = [random.random() for _ in range(1_000_000)]
data_list_b = [random.random() for _ in range(1_000_000)]

# Record time for element-wise addition
t_start = time.time()

sum_list = [a + b for a, b in zip(data_list_a, data_list_b)]

t_end = time.time()

print("Time taken using Python Lists:", t_end - t_start, "seconds")
```

Matrix Multiplication

```python
# Using Python lists
# Build two 1000x1000 matrices
dim = 1000
mat_X = [[random.random() for _ in range(dim)] for _ in range(dim)]
mat_Y = [[random.random() for _ in range(dim)] for _ in range(dim)]

t_begin = time.time()

# Matrix multiplication using nested loops
mat_result = [[0]*dim for _ in range(dim)]

for r in range(dim):
    for c in range(dim):
        for k in range(dim):
            mat_result[r][c] += mat_X[r][k] * mat_Y[k][c]

t_finish = time.time()

print("Matrix Multiplication (List):", t_finish - t_begin, "seconds")
```

```python
# Using NumPy arrays

np_mat_X = np.random.random((1000, 1000))
np_mat_Y = np.random.random((1000, 1000))

t_begin = time.time()

np_mat_result = np.dot(np_mat_X, np_mat_Y)

t_finish = time.time()

print("Matrix Multiplication (NumPy):", t_finish - t_begin, "seconds")
```
